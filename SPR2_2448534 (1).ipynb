```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
f = 5
duration = 1.0

t = np.linspace(0, duration, int(fs*duration), endpoint=False)
x = np.sin(2 * np.pi * f * t)

x
```

```python
plt.figure(figsize=(8,4))
plt.plot(t, x)
plt.title("5 Hz Sine Wave (Time Domain)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
N_dtft = 4096
X_dtft = np.fft.fftshift(np.fft.fft(x, N_dtft))
freq_dtft = np.fft.fftshift(np.fft.fftfreq(N_dtft, 1/fs))

plt.figure(figsize=(8,4))
plt.plot(freq_dtft, np.abs(X_dtft))
plt.title("DTFT (Continuous Spectrum Approximation)")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
N = len(x)
X_dft = np.fft.fftshift(np.fft.fft(x))
freq_dft = np.fft.fftshift(np.fft.fftfreq(N, 1/fs))

plt.figure(figsize=(8,4))
plt.stem(freq_dft, np.abs(X_dft), basefmt=" ")
plt.title("DFT (Discrete Spectrum)")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
f1, f2 = 10, 50  # Frequencies
A1, A2 = 1.0, 0.5

x_comp = A1*np.sin(2*np.pi*f1*t) + A2*np.sin(2*np.pi*f2*t)
x_comp
```

```python
plt.figure(figsize=(8,4))
plt.plot(t, x_comp)
plt.title("Composite Signal (10 Hz + 50 Hz)")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
Xc_dtft = np.fft.fftshift(np.fft.fft(x_comp, N_dtft))
plt.figure(figsize=(8,4))
plt.plot(freq_dtft, np.abs(Xc_dtft))
plt.title("Composite Signal DTFT")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
Xc_dft = np.fft.fftshift(np.fft.fft(x_comp))
plt.figure(figsize=(8,4))
plt.stem(freq_dft, np.abs(Xc_dft), basefmt=" ")
plt.title("Composite Signal DFT")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
alpha = 5  # decay rate
x_exp = np.exp(-alpha * t)
x_exp
```

```python
plt.figure(figsize=(8,4))
plt.plot(t, x_exp)
plt.title("Exponentially Decaying Signal")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
Xe_dtft = np.fft.fftshift(np.fft.fft(x_exp, N_dtft))
plt.figure(figsize=(8,4))
plt.plot(freq_dtft, np.abs(Xe_dtft))
plt.title("Exponential Signal DTFT")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
Xe_dft = np.fft.fftshift(np.fft.fft(x_exp))
plt.figure(figsize=(8,4))
plt.stem(freq_dft, np.abs(Xe_dft), basefmt=" ")
plt.title("Exponential Signal DFT")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
pulse_width = 0.1  # seconds
x_rect = np.zeros_like(t)
x_rect[t < pulse_width] = 1.0
x_rect
```

```python
plt.figure(figsize=(8,4))
plt.plot(t, x_rect)
plt.title("Rectangular Pulse")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
Xr_dtft = np.fft.fftshift(np.fft.fft(x_rect, N_dtft))
plt.figure(figsize=(8,4))
plt.plot(freq_dtft, np.abs(Xr_dtft))
plt.title("Rectangular Pulse DTFT")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

```python
Xr_dft = np.fft.fftshift(np.fft.fft(x_rect))
plt.figure(figsize=(8,4))
plt.stem(freq_dft, np.abs(Xr_dft), basefmt=" ")
plt.title("Rectangular Pulse DFT")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Magnitude")
plt.grid(True)
plt.show()
```

The notebook demonstrates how different types of signals (a single sinusoid, a composite signal of two sinusoids, and a rectangular pulse) have distinct characteristics in the frequency domain.

Single Sinusoid: The frequency domain plots show a single peak at the frequency of the sinusoid, indicating that the signal is composed of only that frequency component.
Composite Signal: The frequency domain plots reveal multiple peaks corresponding to the individual frequencies present in the composite signal. This illustrates how the Fourier transform can decompose a signal into its constituent frequencies.
Rectangular Pulse: The frequency domain plots of the rectangular pulse show a continuous spectrum with a main lobe and side lobes, characteristic of signals with sharp transitions in the time domain.
In conclusion, the notebook visually demonstrates the power of the Fourier transform as a tool for analyzing the frequency content of signals. By transforming a signal from the time domain to the frequency domain, we can gain valuable insights into the underlying frequencies that compose the signal, which is crucial in many signal processing applications.
